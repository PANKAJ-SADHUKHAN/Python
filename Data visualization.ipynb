```python
#Here we see datavisualization usin python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
x=np.random.random(5)
#we use random() to generate random number of 1-d array
y=np.arange(1,6)
#here we use arange() to make an evenly spaced 1-d array element
print(x)
print(y)
print(x.shape)
print(x.dtype)
a=[5.2,4,3,2.6,1.6]
print(type(a))
z=np.array(a)
print(z.dtype)
print(z)
```

```python
plt.scatter(x,y)
plt.scatter(x,z)
plt.title("Demo plot")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["This is y","This is z"])
```

```python
data = pd.read_csv('countries.csv')
```

```python
data
```

```python
#compare the populations of India and Chaina
```

```python
#Boolean Indexing
data.country=='India'
```

```python
print(data[data.country=='India'])
```

```python
India=data[data.country=='India']
print(India)
```

```python
China=data[data.country=='China']
```

```python
US=data[data.country=='United States']
```

```python
plt.plot(India.year,India.population)
plt.show()
```

```python
#here we divide the population with the number 10^6 to get the real value of the population
plt.plot(India.year,India.population/10**6)
plt.plot(China.year,China.population/10**6)
plt.title("India vs China Population")
plt.xlabel("Year")
plt.ylabel("Population")
plt.legend(["India", "China"])
plt.show()
```

```python
#Let's check the growth in percentage
```

```python
India.population
```

```python
plt.plot(India.year,India.population/India.population.iloc[0])
plt.plot(China.year,China.population/China.population.iloc[0])
plt.plot(US.year,US.population/US.population.iloc[0])
plt.title("India vs China vs US Population")
plt.xlabel("Year")
plt.ylabel("Population (First year=100)")
plt.legend(["India", "China", "US"])
plt.show()
```
